```python
RNG = np.random.RandomState(seed=446)

# ns = [20, 200, 500, 1000]
n = 20
vanilla_times = []
numpy_times = []

# for n in ns:
    # Generate some data
x = RNG.randn(n)
y = RNG.randn(n)
A = RNG.randn(n, n)
B = RNG.randn(n, n)
# And their vanilla List equivalents
x_list = x.tolist()
y_list = y.tolist()
A_list = A.tolist()
B_list = B.tolist()
```

```python
n = len(x_list)
f = np.zeros([n])
for j in range(n):
    row = []
    for i in range(n):
        row.append( x_list[i]*(A_list[i][j]+A_list[j][i]) + y_list[i]*B_list[i][j] )
    f[j] = np.sum(row)
f
```
